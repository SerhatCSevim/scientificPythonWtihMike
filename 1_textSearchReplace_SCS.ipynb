```python
import re
```

```python
sentence = "Hi, my name is Serhat and I like to eat Serhat's Pizza."

regular_expression_obj = re.compile('Serhat')
regular_expression_obj.findall(sentence)
```

```python
# Substitute relavent word

regular_expression_obj.sub('SERHAT CAN', sentence)
```

```python
names = ['Adam', 'Bob', 'Charlie']

for name in names:
    print(regular_expression_obj.sub(name,sentence))
```

```python
sentence = "Hi, my name is Serhat and I like to eat Sarhat's Pizza."

regular_expression_obj = re.compile(r'S\wrhat') # \w search for any character

regular_expression_obj.findall(sentence)
```

```python
# pattern to replace
# \n\n1\n00:05.080 --> 00:09.310\n
# \n\n\d\n\d\d:\d\d.\d\d\d --> \d\d:\d\d.\d\d\d\n               \d for numbers

patternToReplace = r'\n\n\d\n\d\d:\d\d.\d\d\d --> \d\d:\d\d.\d\d\d\n'

regex_timestamps = re.compile(patternToReplace)
```

```python
regex_timestamps.findall(text)
```

```python
# Only n0 to n9, work but not completly

patternToReplace = r'\n\n\d{1,3}\n\d\d:\d\d.\d\d\d --> \d\d:\d\d.\d\d\d\n'

regex_timestamps = re.compile(patternToReplace)
regex_timestamps.findall(text)
```

```python
newText = regex_timestamps.sub(' ', text)
newText = newText[7:]
```

```python
words = newText.split()
print(type(words))
print(len(words))
```

```python
for w in range(len(words)):
    if len(words[w]) == 4 and not words[w][-1] == '.':
        words[w] = words[0] + '***'  
print(words)            
```

```python
cleanedText = ' '.join(words)
```

```python
# Download to the computer
with open('cleaned_transcript.txt','w') as f:
    f.write(cleanedText)
```
