```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def nChoosek(n,k):
    num = math.factorial(n) # numerator
    den = math.factorial(k) * math.factorial(n-k) #denominator
    return num/den
```

# Time Dilation in Special Relativity

delta(t') = (delta(t)) / (1 - (v / c)**2) ** 0.5

c = 299,792,458 m/s

```python
# This is tell us the relative dilation of time for someone travelling at velocity of relative to anyone
```

```python
c = 299792458

timedilation = lambda v : 1 / np.sqrt(1 - (v**2 / c**2))

kph = np.zeros(6)

kph[0] = 100
kph[1] =200
kph[2] = 900
kph[3] = 11300
kph[4] = 107991
kph[5] = 828000

labels = ['car', 'autobahn', 'plane', 'satellite', 'Earth', 'Solar System']
```

```python
td = np.zeros(len(kph))

for i in range(len(kph)):
    # convert to kph to m/s
    v = kph[i]*1000 / (60*60)
    
    # compute the time dilation
    td[i] = timedilation(v)
    
td

plt.bar(range(len(td)),td)
plt.ylim(.9999999,1.0000004)
plt.xticks(ticks=range(len(td)),labels=labels, rotation =45)
plt.show()
```

a matrisinin özdeğerlerinin çarpımı, o matrisin determinantına eşittir.
the determinant of a is equal to the product of its eigenvalue."

bir matrisin rankı sıfırdan farklı özdeğer sayısına eşittir.
"the rank of a is equal to the number of nonzero eigenvalues of a."

A square matrix is going to have eigenvalues and eigenvectors and they come in pairs and there 
are M of them. 

Create matrix A and then extract all of the eigenvalues from that matrix and plot those eigenvalues
on a complex plane.

```python
fig, ax = plt.subplots(1,2,figsize=(12,5))

n = 15

for i in range(100):
    # create the random numbers matrix
    A = np.random.randn(n,n)/np.sqrt(n)        # setting standard deviation
    L = np.linalg.eig(A)[0]
    ax[0].plot(np.real(L), np.imag(L), 'o')
    ax[1].plot(np.abs(L), 'o')
    
plt.show()
```

```python
# What we need to do is save all of the eigenvalue magnitudes and then later we can sort them.

fig, ax = plt.subplots(1,2,figsize=(12,5))

n = 15

mags = []

for i in range(100):
    # create the random numbers matrix
    A = np.random.randn(n,n)/np.sqrt(n)        # setting standard deviation
    L = np.linalg.eig(A)[0]
    mags = np.concatenate((mags, np.abs(L)))     # concatenate the magnitudes and the absolute value of the new magnitudes.
    ax[0].plot(np.real(L), np.imag(L), 'o')
    ax[1].plot(np.sort(np.abs(L)), 'o')
    
# We need to do is sort the vector of magnitudes 

mags.sort() 
ax[1].plot(mags[::-1], 'o')
    
plt.show()

# I could not find why second plots seems like that
```
